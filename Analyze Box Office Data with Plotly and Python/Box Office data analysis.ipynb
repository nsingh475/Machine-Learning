# Analyzing Movie Release Dates

```python
test.loc[test.release_date.isna() == False, 'release_date']
```

### 9. Preprocessing Features

```python
def fix_dates(x):
    year = x.split('/')[2]
    if int(year) <=19:
        return x[:-2] + '20' + year
    else:
        return x[:-2] + '19' +year
```

```python
test[test.release_date.isna()]
```

```python
test.loc[test.release_date.isna() == True, 'release_date'] = '05/01/00'
```

```python
train.release_date = train.release_date.apply(lambda x: fix_dates(x))
test.release_date = test.release_date.apply(lambda x: fix_dates(x))
```

### 10. Creating Features Based on Release Date

```python
train.release_date = pd.to_datetime(train.release_date)
test.release_date = pd.to_datetime(test.release_date)
```

```python
def process_dates(df):
    date_parts = ['year', 'weekday', 'month', 'weekofyear', 'day','quarter']
    for part in date_parts:
        part_cols = 'release_date' + '_' +part
        df[part_cols] = getattr(df.release_date.dt, part).astype(int)
    return df

train = process_dates(train)
test = process_dates(test)  
```

### 11. Using Plotly to Visualize the Number of Films Per Year

```python
d1 = train.release_date.value_counts().sort_index()
d2 = test.release_date.value_counts().sort_index()
```

```python
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go

data = [go.Scatter(x=d1.index, y=d1.values, name='train'),
       go.Scatter(x=d2.index, y=d2.values, name='test')]

layout = go.Layout(dict(title='Number of films per year',
                       xaxis=dict(title='Year'),
                       yaxis = dict(title='Count'),
                       ),legend= dict(orientation='v'))

py.iplot(dict(data=data, layout=layout))
```

### 12. Number of Films and Revenue Per Year

```python
d1 = train.release_date.value_counts().sort_index()
d2 = train.groupby(['release_date'])['revenue'].sum()



data = [go.Scatter(x=d1.index, y=d1.values, name='film count'),
       go.Scatter(x=d2.index, y=d2.values, name='toal revenue',yaxis ='y2')]

layout = go.Layout(dict(title='Number of films and revenue per year',
                       xaxis=dict(title='Year'),
                       yaxis = dict(title='Count'),
                       yaxis2 = dict(title='Total revenue', overlaying='y', side='right')
                       ),legend= dict(orientation='v'))

py.iplot(dict(data=data, layout=layout))
```

```python
d1 = train.release_date.value_counts().sort_index()
d2 = train.groupby(['release_date'])['revenue'].mean()



data = [go.Scatter(x=d1.index, y=d1.values, name='film count'),
       go.Scatter(x=d2.index, y=d2.values, name='mean revenue',yaxis ='y2')]

layout = go.Layout(dict(title='Number of films and revenue per year',
                       xaxis=dict(title='Year'),
                       yaxis = dict(title='Count'),
                       yaxis2 = dict(title='mean revenue', overlaying='y', side='right')
                       ),legend= dict(orientation='v'))

py.iplot(dict(data=data, layout=layout))
```
